```python
import os
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

directory=r"E:\Study materials\Artificial intelligence\Research work\Breast cancer Conference\Dataset\Dataset_BUSI\Breast us dataset"
categories=['benign']
data=[]
for category in categories:
    path=os.path.join(directory,category)
    index=categories.index(category)
    for images in os.listdir(path):
        img_array=cv.imread(os.path.join(path,images))
        img_array=cv.resize(img_array,(500,500))
        data.append(img_array)
data=np.array(data)       
```

```python
from keras.preprocessing.image import ImageDataGenerator
datagen=ImageDataGenerator(
    horizontal_flip=True,
    #transforms.Resize(400)
    rotation_range=20,
    zoom_range=0.2,
    #shear_range=0.1,
    #smart_resize=True,
    fill_mode='reflect'
)
```

```python
i=0
for batch in datagen.flow(data,batch_size=1,
                         save_to_dir='dog1',
                          save_prefix='d3',
                          save_format='png'):
    i+=1
    if i>999:
        break
```
